### Descripción:
Script para la conversión de Pesos Colombianos (COP) a Dólares (USD) y Euros (EUR) utilizando la TRM del día, facilitando la toma de decisiones financieras.

```python
# --- SECCIÓN DE ENTRADA DE DATOS ---
# Solicitamos el monto en pesos. Usamos float() porque el dinero puede tener decimales.
# Vimos esto en la celda 8 de tu clase 2 (float("3.14")).
pesos_colombianos = float(input("Ingrese el valor en Pesos Colombianos (COP) a convertir: "))

# Solicitamos la Tasa Representativa del Mercado (TRM) para el Dólar.
tasa_dolar = float(input("Ingrese la TRM del día para el Dólar (USD): "))

# Solicitamos la Tasa de cambio para el Euro.
tasa_euro = float(input("Ingrese la tasa de cambio del día para el Euro (EUR): "))

# --- SECCIÓN DE CÁLCULOS ---
# Para convertir de una moneda local a una extranjera, dividimos el monto local por la tasa de cambio.
dolares = pesos_colombianos / tasa_dolar
euros = pesos_colombianos / tasa_euro

# --- SECCIÓN DE SALIDA (IMPRESIÓN) ---
# Usamos f-strings para insertar las variables directamente en el texto.
# El código :.2f sirve para limitar a 2 decimales, igual que en tu ejercicio de "Valor Futuro" (Celda 18, clase 2).
print("-" * 40) # Línea decorativa
print(f"RESUMEN FINANCIERO DE CONVERSIÓN")
print("-" * 40)
print(f"Monto inicial:   $ {pesos_colombianos:,.2f} COP")
print(f"Tasa USD usada:  $ {tasa_dolar:,.2f}")
print(f"Tasa EUR usada:  $ {tasa_euro:,.2f}")
print("-" * 40)
print(f"Equivalente USD: $ {dolares:,.2f} Dólares")
print(f"Equivalente EUR: $ {euros:,.2f} Euros")
print("-" * 40)
```
