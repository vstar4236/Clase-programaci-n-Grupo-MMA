# Pseudocódigo: Motor Lógico "FinVes"

1. INICIO DEL SISTEMA
2. CONFIGURAR PERFIL DE USUARIO

   2.1 INTENTAR LEER nombre del usuario y meta de libertad financiera mensual.

   2.2 SI el usuario ingresa texto en lugar de números para la meta, capturar el error y asignar una meta por defecto de $5,000,000.

   2.3 CREAR DICCIONARIO `perfil_usuario` y GUARDAR nombre y meta.

3. CREAR TUPLA `activos_disponibles` con: "Bancolombia", "Ecopetrol", "S&P500", "Bonos del Estado".
4. CREAR LISTA vacía `portafolio`.
5. MIENTRAS el programa esté activo (Ciclo Infinito):
   5.1 MOSTRAR menú interactivo (1. Registrar, 2. Ver Estado, 3. Salir).

   5.2 LEER opción seleccionada.

   5.3 SI opción es 1 (Registrar):

       5.3.1 MOSTRAR `activos_disponibles`.

       5.3.2 LEER activo elegido.

       5.3.3 INTENTAR LEER monto a invertir y tasa anual esperada.

       5.3.4 SI hay un error de formato (letras en vez de números), MOSTRAR advertencia.

       5.3.5 SI los datos son correctos, GUARDAR inversión estructurada en la lista `portafolio`.

   5.4 SI opción es 2 (Ver Estado):

       5.4.1 SI `portafolio` está vacío:

             5.4.1.1 MOSTRAR advertencia.

       5.4.2 SI NO:

             5.4.2.1 INICIALIZAR ingreso_pasivo_total en 0.

             5.4.2.2 PARA CADA inversión en `portafolio`:

                     5.4.2.2.1 CALCULAR pasivo_mensual = (Monto * (Tasa / 100)) / 12.

                     5.4.2.2.2 SUMAR pasivo_mensual al ingreso_pasivo_total.

             5.4.2.3 SI ingreso_pasivo_total >= meta de `perfil_usuario`:

                     5.4.2.3.1 IMPRIMIR mensaje de éxito total de Libertad Financiera.

             5.4.2.4 SI NO:

                     5.4.2.4.1 CALCULAR porcentaje_avance = (ingreso_pasivo_total / meta) * 100.

                     5.4.2.4.2 IMPRIMIR mensaje indicando el porcentaje de avance.

   5.5 SI opción es 3 (Salir):

       5.5.1 TERMINAR el programa.
       
6. FIN


```python
# ==========================================
# MOTOR LÓGICO FINVES - LIBERTAD FINANCIERA
# ==========================================

print("🚀 BIENVENIDO A FINVES - TU CAMINO A LA LIBERTAD FINANCIERA 🚀\n")

# PASO 1: Configuración de Estructuras de Datos
try:
    # EJEMPLO DE ENTRADA: Matías
    nombre_input = input("Ingresa tu nombre (Ejemplo: Matías): ").strip()
    
    # EJEMPLO DE ENTRADA: 5000000
    meta_input = float(input("Ingresa tu Meta de Libertad Financiera mensual en COP (Ejemplo: 5000000 para 5 millones): "))
except ValueError:
    print("⚠️ Error: Debes ingresar un número válido para la meta. Se asignará $5,000,000 por defecto.")
    meta_input = 5000000.0

# 1.1 Diccionario
perfil_usuario = {
    "nombre": nombre_input,
    "meta": meta_input
}

# 1.2 Tupla
activos_disponibles = ("Bancolombia", "Ecopetrol", "S&P500", "Bonos del Estado")

# 1.3 Lista
portafolio = []

# PASO 2: Menú Interactivo (while)
while True:
    print("\n" + "="*45)
    print("📊 MENÚ PRINCIPAL FINVES 📊")
    print("="*45)
    print("1. Registrar nueva inversión")
    print("2. Ver Estado del Portafolio y Libertad Financiera")
    print("3. Salir de la App")
    
    # EJEMPLO DE ENTRADA: 1
    opcion = input("\nSelecciona una opción (1, 2 o 3): ").strip()
    
    # PASO 3: Lógica de la Opción 1
    if opcion == "1":
        print("\n--- MERCADOS AUTORIZADOS ---")
        # Mostramos la tupla de activos al usuario
        for i, activo in enumerate(activos_disponibles):
            print(f"- {activo}")
            
        # EJEMPLO DE ENTRADA: S&P500
        activo_elegido = input("\nEscribe el nombre del activo (Ejemplo: S&P500): ").strip()
        
        # Uso de try-except (Bono de +1.0 unidad) para evitar cierres abruptos
        try:
            # EJEMPLO DE ENTRADA: 150000000 (150 millones)
            monto = float(input("Ingresa el monto a invertir en COP (Ejemplo: 150000000 sin puntos ni comas): "))
            
            # EJEMPLO DE ENTRADA: 10 (para un 10%)
            tasa = float(input("Ingresa la Tasa de Rendimiento Anual esperada (Ejemplo: 10): "))
            
            # Guardamos la información estructurada como un diccionario dentro de la lista
            nueva_inversion = {
                "activo": activo_elegido,
                "monto": monto,
                "tasa": tasa
            }
            portafolio.append(nueva_inversion)
            print("✅ ¡Inversión registrada con éxito!")
            
        except ValueError:
            print("❌ Error de formato: Debes ingresar números válidos para el monto y la tasa. Intenta de nuevo.")

    # PASO 4: Lógica de la Opción 2
    elif opcion == "2":
        print("\n--- ESTADO DEL PORTAFOLIO ---")
        
        # Validación si la lista está vacía
        if len(portafolio) == 0:
            print("⚠️ Aún no tienes inversiones registradas. ¡Anímate a dar el primer paso!")
        else:
            ingreso_pasivo_total = 0.0
            
            # Ciclo for para recorrer la lista y calcular
            for inversion in portafolio:
                monto_inv = inversion["monto"]
                tasa_inv = inversion["tasa"]
                
                # Fórmula matemática exigida
                pasivo_mensual = (monto_inv * (tasa_inv / 100)) / 12
                ingreso_pasivo_total += pasivo_mensual
                
                print(f"🔸 {inversion['activo']}: Generando ${pasivo_mensual:,.2f} mensuales.")
            
            print("-" * 45)
            print(f"💰 Ingreso Pasivo Mensual Total: ${ingreso_pasivo_total:,.2f}")
            print(f"🎯 Tu Meta Mensual: ${perfil_usuario['meta']:,.2f}")
            print("-" * 45)
            
            # Decisión lógica (if/else) comparando contra la meta del diccionario
            if ingreso_pasivo_total >= perfil_usuario["meta"]:
                print(f"🎉 ¡Felicidades {perfil_usuario['nombre']}! Has alcanzado la Libertad Financiera.")
            else:
                # Cálculo de porcentaje
                porcentaje_avance = (ingreso_pasivo_total / perfil_usuario["meta"]) * 100
                print(f"🚀 Te falta un camino por recorrer. Tu avance es del {porcentaje_avance:.2f}%.")

    # Salida del sistema
    elif opcion == "3":
        print(f"\nSaliendo de FinVes... ¡Sigue invirtiendo inteligentemente, {perfil_usuario['nombre']}!")
        break
        
    else:
        print("❌ Opción no válida. Por favor digita 1, 2 o 3.")
```
