# Ejercicio Valor Futuro Python

```python
# Solicitud de datos al usuario transformados a int
principal = int(input("Introduce el monto inicial invertido (P): "))
tasa_interes = int(input("Introduce la tasa de interés anual (%) (r): "))
años_n = int(input("Introduce el número de años (t): "))
 
# Cálculo del Valor Futuro (Interés Simple)
# Dividimos la tasa entre 100 para convertir el número entero a decimal
valor_futuro = principal * (1 + (tasa_interes / 100) * años_n)
 
print("-" * 30)
print(f"Al final de {años_n} años, tu inversión valdrá: {valor_futuro:.2f}")
```
