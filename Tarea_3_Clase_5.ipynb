## 📝 FASE DE PLANEACIÓN: PSEUDOCÓDIGO
**Objetivo:** Diseñar la estructura lógica de la clase `CarritoDeCompras` antes de su implementación en código.

1. **DEFINIR CLASE** `CarritoDeCompras`:
   1.1. **Constructor (Estado Inicial):**
        - INICIALIZAR atributo `items` como una lista vacía `[]`.
        - INICIALIZAR atributo `total` en `0`.
        
   1.2. **MÉTODO** `agregar_producto(nombre, precio)`:
        - RECIBIR el nombre y el precio del producto.
        - CREAR una estructura (diccionario) con estos dos datos.
        - AGREGAR el diccionario a la lista `self.items`.
        - SUMAR el precio al atributo `self.total`.
        - IMPRIMIR mensaje confirmando que el producto fue agregado.
        
   1.3. **MÉTODO** `ver_carrito()`:
        - IMPRIMIR encabezado del recibo.
        - SI la lista `items` está vacía, IMPRIMIR advertencia.
        - SINO, PARA CADA producto en la lista `items`:
            - IMPRIMIR el nombre del producto y su precio.
        - IMPRIMIR el costo total acumulado `self.total`.

2. **FASE DE PRUEBA (Instanciación):**
   2.1. CREAR el objeto `mi_compra` a partir de la clase `CarritoDeCompras`.
   2.2. LLAMAR al método `agregar_producto` 3 veces con diferentes artículos.
   2.3. LLAMAR al método `ver_carrito` para mostrar el resumen final.

```python
# ==============================================================================
# IMPLEMENTACIÓN DE LA CLASE: CARRITO DE COMPRAS (POO)
# ==============================================================================

# Definimos la plantilla principal (Clase) que representará nuestro carrito
class CarritoDeCompras:
    
    # El método __init__ es el constructor. Se ejecuta automáticamente al crear un objeto.
    def __init__(self):
        # self.items es un atributo de instancia que guarda los productos (inicia vacío)
        self.items = []
        # self.total es un atributo de instancia que acumula el costo (inicia en cero)
        self.total = 0
        
    # Método para añadir un nuevo artículo al carrito
    def agregar_producto(self, nombre_producto, precio_producto):
        # Creamos un diccionario temporal con los datos del producto ingresado
        nuevo_item = {"producto": nombre_producto, "precio": precio_producto}
        
        # Agregamos (adjuntamos) ese diccionario a nuestra lista de items del carrito
        self.items.append(nuevo_item)
        
        # Sumamos el precio del nuevo producto al total acumulado del carrito
        self.total += precio_producto
        
        # Imprimimos un mensaje de confirmación para mejorar la experiencia del usuario
        print(f"✅ Se agregó '{nombre_producto}' al carrito. (Precio: ${precio_producto:,.2f})")
        
    # Método para visualizar el resumen de la compra
    def ver_carrito(self):
        # Imprimimos un separador visual para el encabezado del recibo
        print("\n" + "="*40)
        print("🧾 RESUMEN DE TU CARRITO DE COMPRAS")
        print("="*40)
        
        # Validamos si el carrito tiene elementos antes de intentar mostrarlos
        if len(self.items) == 0:
            # Mensaje por si el cliente intenta ver un carrito vacío
            print("Tu carrito está vacío.")
        else:
            # Ciclo for para recorrer cada producto guardado en la lista de items
            for item in self.items:
                # Imprimimos el nombre y el precio de cada artículo formateado
                print(f" 🛒 {item['producto']} ..... ${item['precio']:,.2f}")
                
        # Imprimimos el separador final del recibo
        print("-" * 40)
        # Mostramos el costo total acumulado que fuimos calculando
        print(f"💰 TOTAL A PAGAR: ${self.total:,.2f}")
        print("=" * 40 + "\n")

# ==============================================================================
# PRUEBA DE LA CLASE (INSTANCIACIÓN Y USO)
# ==============================================================================

# 1. Instanciamos la clase: Creamos un objeto real llamado 'mi_compra'
mi_compra = CarritoDeCompras()

# 2. Usamos el método agregar_producto() para insertar 3 productos del catálogo de repuestos
mi_compra.agregar_producto("Kit de Arrastre Ketoz Racing", 125000)
mi_compra.agregar_producto("Disco de Freno Delantero", 85000)
mi_compra.agregar_producto("Juego de Pastillas de Freno", 45000)

# 3. Usamos el método ver_carrito() para validar que la lógica interna sumó y guardó todo bien
mi_compra.ver_carrito()
```
