```python
# El if y el else
# El programa evalúa una condición. Si es Verdadera, actúa. Si es Falsa, toma otro camino.

# Sintaxis y Ejemplo (Finanzas - Riesgo Crediticio):

score_credito = 750

if score_credito >= 700:
    print("Crédito Aprobado: Tasa Preferencial")
else:
    print("Crédito Rechazado: Riesgo Alto")

# La indentación (sangría) es obligatoria. Es la forma en que Python sabe qué código pertenece al if y cuál está fuera.
```

```python
# Decisiones Múltiples (elif)

# Ejemplo (Logística - Costos de Envío):

peso_paquete = float(input("Ingrese el peso del paquete en KG: "))

if peso_paquete < 2:
    costo = 5000
    tipo = "Sobre"
elif peso_paquete < 20:
    costo = 20000
    tipo = "Caja Estándar"
else:
    costo = 50000
    tipo = "Carga Pesada"

print(f"Tipo: {tipo} | Costo: ${costo}")
```

```python
"""Operadores Lógicos (and, or)

Refinando la estrategia.

    and (Y): Ambas condiciones deben ser verdaderas. (Exigente).

    or (O): Basta con que una sea verdadera. (Flexible)."""

# Ejemplo (Mercadeo - Segmentación):

edad = 25
suscripcion = "Premium"

# Campaña para jóvenes VIP
if edad < 30 and suscripcion == "Premium":
    print("Enviar cupón del 20% de descuento")

# Campaña de reactivación general
if suscripcion == "Inactivo" or suscripcion == "Básico":
    print("Enviar correo de 'Te extrañamos'")
```

```python
""" Sentencia match (Python 3.10+)

La evolución moderna del if/elif para menús y categorías.

Ideal para cuando comparamos una misma variable contra múltiples valores fijos."""

# Ejemplo (Menú de Usuario):

opcion = input("Seleccione perfil: (1) Admin, (2) Vendedor, (3) Cliente: ")

match opcion:
    case "1":
        print("Acceso total al sistema.")
    case "2":
        print("Acceso al módulo de ventas.")
    case "3":
        print("Acceso al catálogo.")
    case _:
        print("Opción no válida.") # El guion bajo (_) es el "default"
```

```python
"""CICLOS INDEFINIDOS (WHILE)

El Bucle while

Repetir MIENTRAS una condición sea verdadera.

Peligro: Si la condición nunca se vuelve falsa, creamos un Bucle Infinito (el programa se cuelga). Debemos garantizar que la variable de control cambie."""

# Ejemplo (Gestión de Inventario):

stock = 10
ventas = 0

while stock > 0:
    print(f"Vendiendo producto... Stock actual: {stock}")
    stock -= 1  # ¡Importante! Reducir el stock
    ventas += 1

print(f"¡Sold Out! Inventario agotado. Total de ventas: {ventas}")
```

```python
"""Control de Emergencia (break y continue)

Tomando el control manual del ciclo.

    break: Rompe el ciclo y sale inmediatamente. (Abortar misión).

    continue: Salta la vuelta actual y vuelve al inicio del ciclo. (Omitir paso).
"""

# Ejemplo (Búsqueda en Base de Datos):

contador = 0
while contador < 10:
    contador += 1
    
    if contador == 5:
        continue  # Se salta el 5 y sigue con el 6
    
    if contador == 8:
        break     # Se detiene totalmente en el 8
        
    print(f"Procesando cliente #{contador}")
```

```python

# Ejemplo (Logística - Costos de Envío):

try:
    while True:
        entrada = input("Ingrese el peso del paquete en KG: ").strip()
        if entrada == "":
            print("Error: el campo no puede estar vacío. Intente nuevamente.")
            continue
        try:
            peso = float(entrada)
        except ValueError:
            print("Error: debe ingresar un número (ej. 1.5). Intente nuevamente.")
            continue
        if peso <= 0:
            print("Error: el peso debe ser mayor que 0. Intente nuevamente.")
            continue
        peso_paquete = peso
        break
except (KeyboardInterrupt, EOFError):
    print("\nEntrada cancelada. Se mantiene el peso actual:", peso_paquete)

if peso_paquete < 2:
    costo = 5000
    tipo = "Sobre"
elif peso_paquete < 20:
    costo = 20000
    tipo = "Caja Estándar"
else:
    costo = 50000
    tipo = "Carga Pesada"

print(f"Tipo: {tipo} | Costo: ${costo}")
```

```python
# Ejercicio 1 (Mercadeo): El Segmentador

# Solicita la edad al usuario y elimina espacios al inicio y al final
edad_input = input("Ingrese su edad (años): ").strip()

# Intenta convertir la entrada de edad a entero; si falla, muestra error y termina
try:
    edad = int(edad_input)  # convierte la edad a entero para comparaciones
    if edad <= 0:  # valida que la edad sea un número positivo
        print("Error: la edad debe ser un número mayor que 0.")
        raise SystemExit  # termina el script por entrada inválida
except ValueError:  # captura si la conversión a entero falla
    print("Error: ingrese un número entero para la edad.")
    raise SystemExit  # termina el script por entrada inválida

# Solicita los ingresos al usuario y elimina espacios
ingresos_input = input("Ingrese sus ingresos anuales (ej. 2500000 o $2.500.000): ").strip()
# Normaliza la cadena de ingresos quitando símbolos comunes como $ y separadores de miles
ingresos_limpio = ingresos_input.replace("$", "").replace(" ", "").replace(",", "").replace(".", "")

# Intenta convertir los ingresos a número; si falla, muestra error y termina
try:
    ingresos = float(ingresos_limpio)  # convierte a float para permitir decimales
    if ingresos < 0:  # valida que los ingresos no sean negativos
        print("Error: los ingresos no pueden ser negativos.")
        raise SystemExit  # termina el script por entrada inválida
except ValueError:  # captura si la conversión a float falla
    print("Error: ingrese un número válido para los ingresos.")
    raise SystemExit  # termina el script por entrada inválida

# Define los umbrales en valores numéricos (2M = 2_000_000 ; 5M = 5_000_000)
umbral_joven = 2_000_000  # ingresos mínimos para el segmento joven potencial
umbral_oro = 5_000_000    # ingresos mínimos para el segmento oro

# Aplica la lógica de segmentación según la edad y los ingresos
if edad < 25 and ingresos > umbral_joven:
    print("Segmento Joven Potencial")  # caso: joven con ingresos altos relativos
elif edad > 25 and ingresos > umbral_oro:
    print("Segmento Oro")  # caso: mayor de 25 con ingresos muy altos
else:
    print("Segmento Estándar")  # cualquier otro caso cae en segmento estándar
```

```python
# Ejercicio 2 (Logística): Control de Calidad (while)

# Simulación de una banda transportadora con 20 productos (ejercicio de control de calidad).
producto_num = 1  # número del producto actual en la banda, inicia en 1
max_productos = 20  # cantidad total de productos que pasan por la banda
empaquetados = 0  # contador de productos que se empaquetan correctamente

while producto_num <= max_productos:  # repetir mientras queden productos por procesar
    print(f"Procesando producto #{producto_num}")  # mostrar qué producto se está procesando
    
    if producto_num == 13:  # si el producto es el número 13, está defectuoso
        print(" -> 🔴 Producto defectuoso: no se empaca.")  # informar que se omitirá el empaque
        producto_num += 1  # avanzar al siguiente producto antes de continuar
        continue  # saltar el resto del cuerpo del while y pasar a la siguiente iteración
    
    if producto_num == 18:  # si el producto es el número 18, la máquina falla
        print(" -> ❌ Falla en la máquina: detener la producción.")  # informar la falla
        break  # salir inmediatamente del bucle por la falla
    
    print(" -> Producto empaquetado correctamente.")  # informar que el producto fue empaquetado
    empaquetados += 1  # aumentar el contador de empaquetados
    producto_num += 1  # pasar al siguiente producto en la banda

print(f"\n Producción finalizada. Total empaquetados: {empaquetados}")  # resumen final de la simulación
```

```python
# Ejercicio 3 (Administración): Menú Interactivo (match)

# Inicializa la variable que guardará la opción elegida por el usuario (cadena vacía para empezar).
opcion = ""

# Bucle que repite el menú hasta que el usuario elija salir.
while True:
    # Muestra el menú de opciones al usuario.
    print("\n¿Desea (1) Calcular IVA, (2) Calcular Descuento, (3) Salir?")
    # Lee la opción ingresada por el usuario y elimina espacios al inicio y al final.
    opcion = input("Ingrese (1) Calcular IVA, (2) Calcular Descuento, (3) Salir: ").strip() # strip() para limpiar la entrada de espacios innecesarios.
    
    # Usa 'match' para decidir qué acción ejecutar según la opción.
    match opcion:
        # Caso 1: calcular IVA
        case "1":
            # Pide al usuario el monto base sobre el cual calcular el IVA.
            monto_input = input("Ingrese el monto sin IVA (ej. 1000.50): ").strip()
            # Intenta convertir la entrada a número flotante; si falla, muestra error y vuelve al menú.

            try:
                monto = float(monto_input)
            except ValueError:
                print("Entrada inválida: ingrese un número válido para el monto.")
                continue  # Vuelve al inicio del while para mostrar el menú otra vez.

            # Define la tasa de IVA (ejemplo: 19%).
            tasa_iva = 0.19
            # Calcula el IVA multiplicando el monto por la tasa.
            iva = monto * tasa_iva
            # Calcula el total sumando el monto y el IVA.
            total_con_iva = monto + iva
            # Muestra el resultado al usuario con formato legible.
            print(f"Monto: {monto:.2f} | IVA ({tasa_iva*100:.0f}%): {iva:.2f} | Total: {total_con_iva:.2f}")
        
        # Caso 2: calcular descuento
        case "2":
            # Pide al usuario el precio original del producto o servicio.
            precio_input = input("Ingrese el precio original (ej. 1000.50): ").strip()
            # Pide al usuario el porcentaje de descuento a aplicar.
            porcentaje_input = input("Ingrese el porcentaje de descuento (ej. 20 para 20%): ").strip()
            # Intenta convertir ambas entradas a números; si falla, informa y vuelve al menú.

            try:
                precio = float(precio_input)
                porcentaje = float(porcentaje_input)
            except ValueError:
                print("Entrada inválida: ingrese números válidos para precio y porcentaje.")
                continue  # Vuelve al inicio del while.

            # Valida que el porcentaje esté en un rango razonable (0 a 100).
            if porcentaje < 0 or porcentaje > 100:
                print("Porcentaje inválido: debe estar entre 0 y 100.")
                continue  # Vuelve al inicio del while.

            # Calcula el monto del descuento.
            monto_descuento = precio * (porcentaje / 100)
            # Calcula el precio final restando el descuento al precio original.
            precio_final = precio - monto_descuento

            # Muestra el resultado al usuario con formato.
            print(f"Precio original: {precio:.2f} | Descuento: {monto_descuento:.2f} ({porcentaje:.1f}%) | Precio final: {precio_final:.2f}")
        
        # Caso 3: salir del menú
        case "3":
            # Informa que se cierra la aplicación y rompe el bucle.
            print("Saliendo... ¡Hasta luego!")
            break  # Sale del while y termina el menú.
        
        # Caso por defecto: opción no válida
        case _:
            # Informa al usuario que la opción ingresada no es válida y repite el menú.
            print("Opción no válida. Por favor ingrese 1, 2 o 3.")
```
