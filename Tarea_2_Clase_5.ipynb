```python
# ==============================================================================
# DATASET: VENTAS SEMANALES
# ==============================================================================
# Esta lista representa las ventas diarias de Lunes a Domingo.
ventas_semanales = [1250000, 1800000, 950000, 2100000, 1600000, 750000, 2300000]
```

# 1️⃣ CÁLCULO DEL TOTAL DE VENTAS

**Prompt enviado a la IA (XML):**

Actúa como un analista de datos. Escribe un código en Python que calcule la suma total de los valores en la lista 'ventas_semanales'. Después de mostrar el código, explica brevemente cómo funciona la función utilizada para sumar.

**Explicación de la IA:**
El código utiliza la función integrada `sum()` de Python. Esta función toma un elemento iterable (como nuestra lista de ventas) y recorre cada uno de sus elementos sumándolos internamente de forma automática, sin necesidad de escribir un ciclo `for` manual. # 2️⃣ CÁLCULO DEL PROMEDIO DIARIO

**Prompt enviado a la IA (XML):**
<instruccion>
Genera el código en Python para calcular el promedio de venta diario de la lista 'ventas_semanales'. Explica matemáticamente y en código cómo obtuviste el promedio utilizando las funciones `sum()` y `len()`.
</instruccion>

**Explicación de la IA:**
El promedio (o media aritmética) se calcula dividiendo la suma total de los elementos entre la cantidad de elementos. En Python, usamos `sum(ventas_semanales)` para obtener el monto total vendido y `len(ventas_semanales)` para contar cuántos días (elementos) hay en la lista. Al dividir ambos valores, obtenemos el promedio diario.

```python
# Cálculo del promedio de ventas
promedio_diario = sum(ventas_semanales) / len(ventas_semanales)
print(f"📊 El promedio de venta diario es: ${promedio_diario:,.2f} COP")
```

# 3️⃣ CONTEO DE DÍAS QUE SUPERARON LA META ($1,500,000)

**Prompt enviado a la IA (XML):**
<instruccion>
Escribe un script en Python que cuente cuántos días de la semana tuvieron ventas estrictamente mayores a 1500000 COP en la lista 'ventas_semanales'. Explica la lógica del ciclo 'for' y el uso de la variable contador en este contexto.
</instruccion>

**Explicación de la IA:**
Para resolver esto, se implementa un patrón de "conteo". Se inicializa una variable `dias_exitosos` en 0. Luego, un ciclo `for` recorre cada valor de la lista. Dentro del ciclo, un condicional `if` evalúa si la venta de ese día es mayor a 1,500,000. Si la condición se cumple, la variable contador aumenta en 1 (`+= 1`). Al finalizar el ciclo, el contador tendrá el número exacto de días que superaron la meta.

```python
# Conteo de días con ventas superiores a 1.5M
dias_exitosos = 0
meta = 1500000

for venta in ventas_semanales:
    if venta > meta:
        dias_exitosos += 1

print(f"🎯 Días en los que la venta superó los $1,500,000 COP: {dias_exitosos} días")
```

# 4️⃣ DÍA DE LA SEMANA CON LA VENTA MÁS ALTA

**Prompt enviado a la IA (XML):**
<instruccion>
Teniendo en cuenta que los datos de 'ventas_semanales' corresponden cronológicamente a los días de lunes a domingo. Escribe un código en Python que encuentre el valor máximo de venta y el nombre del día de la semana correspondiente. Explica cómo relacionaste el índice del valor máximo con los días de la semana.
</instruccion>

**Explicación de la IA:**
Para lograr esto, primero definimos una nueva lista llamada `dias` que contiene los nombres de lunes a domingo en el mismo orden cronológico. Usamos la función `max()` para encontrar el valor más alto en `ventas_semanales`. Luego, usamos el método `.index()` para averiguar en qué posición (índice 0 a 6) se encuentra ese valor máximo. Como ambas listas están sincronizadas, usamos ese mismo índice para extraer el nombre del día correcto de la lista `dias`.

```python
# Identificación del mejor día de ventas
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

# Encontramos el valor más alto
venta_maxima = max(ventas_semanales)

# Encontramos la posición (índice) de ese valor
indice_maximo = ventas_semanales.index(venta_maxima)

# Usamos el índice para encontrar el día correspondiente
dia_estrategico = dias[indice_maximo]

print(f"🏆 El día con la venta más alta fue el {dia_estrategico} con un total de ${venta_maxima:,.2f} COP")
```
