```python
# Ejercicio: El "Organizador de Precios"
# Script para ordenar listas de precios y calcular el IVA automáticamente.

# 1. Definimos la lista inicial desordenada de precios.
precios = [100, 20, 50, 80, 10]

# 2. Utilizamos el método .sort() para ordenar la lista de menor a mayor.
precios.sort()

# 3. Creamos una lista vacía donde guardaremos los nuevos precios calculados.
precios_con_iva = []
tasa_iva = 0.19 # Definimos el IVA al 19%.

# Usamos un bucle 'for' para recorrer cada precio en nuestra lista ya ordenada.
for precio in precios:
    # Calculamos el impuesto y el total, similar al menú interactivo de la clase pasada.
    iva = precio * tasa_iva
    precio_total = precio + iva
    
    # Agregamos el precio final a nuestra nueva lista usando el método .append()
    precios_con_iva.append(precio_total)

# 4. Imprimimos ambas listas para evidenciar la diferencia entre modificar una lista y generar una nueva.
print(f"Precios base (ordenados): {precios}")
print(f"Precios finales (con 19% IVA): {precios_con_iva}")
```
