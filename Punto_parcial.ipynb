```python
# Tarea Clase 3: Sistema de Login Seguro

# 1. Defina una contraseña_correcta en una variable
contrasena_correcta = "MiClaveSegura" 

intentos = 0
max_intentos = 3

# 2. Use un while para permitir máximo 3 intentos
while intentos < max_intentos:
    # 3. Pida la contraseña al usuario (input)
    intento_usuario = input("Ingrese su contraseña bancaria: ")
    intentos += 1
    
    if intento_usuario == contrasena_correcta:
        # 4. Si acierta -> print("Bienvenido") y break
        print("Bienvenido")
        break
    else:
        intentos_restantes = max_intentos - intentos
        
        if intentos_restantes == 0:
            # 6. Si agota los intentos -> print("Cuenta Bloqueada")
            print("Cuenta Bloqueada")
        else:
            # 5. Si falla -> Muestre intentos restantes
            print(f"Contraseña incorrecta. Intentos restantes: {intentos_restantes}")
```
