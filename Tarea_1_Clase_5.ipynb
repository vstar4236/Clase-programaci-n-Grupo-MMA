```python
# ANALIZANDO UNA CAMPAÑA DE MARKETING CON IA
# 1. DATASET DE TRABAJO

# Esta es la base de datos de la campaña que analizaremos:
datos_campania = [
    {"cliente_id": "C001", "ciudad": "Bogotá", "gasto": 150000},
    {"cliente_id": "C002", "ciudad": "Medellín", "gasto": 85000},
    {"cliente_id": "C003", "ciudad": "Bogotá", "gasto": 220000},
    {"cliente_id": "C004", "ciudad": "Cali", "gasto": 110000},
    {"cliente_id": "C005", "ciudad": "Medellín", "gasto": 180000},
]
```

**Respuesta generada por la IA:**
La variable `datos_campania` es una **lista de diccionarios** en Python. 
* **La Lista (`[]`):** Funciona como el contenedor principal que agrupa todos los registros de la campaña.
* **Los Diccionarios (`{}`):** Cada diccionario representa un registro individual (un consumidor específico).
* **Las Claves (Keys):**
    * `cliente_id`: Es el identificador único (cadena de texto) asignado a cada usuario para mantener el anonimato en la base de datos.
    * `ciudad`: Indica la ubicación geográfica de la transacción, un dato fundamental para segmentar la campaña por regiones.
    * `gasto`: Representa el valor monetario (entero) que el cliente invirtió, clave para medir el Retorno de Inversión (ROI) y detectar clientes de alto valor.

```python
# ==============================================================================
# EJERCICIO 2: PROMPT 2 (CÁLCULO TOTAL)
# ==============================================================================

# MI PROMPT PARA LA IA:
# "Genera el código en Python para calcular el gasto total de la campaña sumando el valor de la clave 'gasto' de todos los clientes en la lista 'datos_campania'. Usa un ciclo for, documenta cada paso con comentarios y muestra el resultado final con formato de moneda."

# --- CÓDIGO GENERADO POR LA IA ---

# 1. Inicializamos la variable acumuladora en cero
gasto_total = 0

# 2. Recorremos cada diccionario (cliente) dentro de la lista principal
for cliente in datos_campania:
    # 3. Extraemos el valor de la clave 'gasto' y lo sumamos al acumulador
    gasto_total += cliente["gasto"]

# 4. Imprimimos el resultado usando f-strings para separar los miles
print(f"💰 El gasto total de la campaña de marketing es: ${gasto_total:,.2f}")
```

```python
# ==============================================================================
# EJERCICIO 3: PROMPT 3 (FILTRADO Y ANÁLISIS)
# ==============================================================================

# MI PROMPT PARA LA IA:
# "Genera el código en Python para filtrar la lista 'datos_campania'. Usa un ciclo for y condicionales lógicos para imprimir únicamente a los clientes que cumplan DOS condiciones simultáneamente: que su ciudad sea 'Bogotá' AND que su gasto sea estrictamente mayor a 200000. Agrega comentarios explicativos."

# --- CÓDIGO GENERADO POR LA IA ---

print("🔍 Clientes de Alto Valor en Bogotá (Gasto > $200,000):")
print("-" * 55)

# 1. Iteramos sobre cada cliente en la base de datos
for cliente in datos_campania:
    
    # 2. Validamos con el operador 'and' que se cumplan ambas condiciones obligatorias
    if cliente["ciudad"] == "Bogotá" and cliente["gasto"] > 200000:
        
        # 3. Si el cliente cumple, mostramos su información formateada
        print(f"✅ Cliente ID: {cliente['cliente_id']} | Gasto Registrado: ${cliente['gasto']:,.2f}")
```
