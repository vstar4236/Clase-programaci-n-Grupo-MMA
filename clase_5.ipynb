```python
# ===============================================================================
# 1. BIENVENIDA E INTERACCIÓN INICIAL
# ===============================================================================

# Mostrar un título atractivo con emojis para captar la atención del usuario
print("=" * 70)
print("💰 CALCULADORA FINANCIERA DE DIVISAS 💱")
print("=" * 70)

# Mostrar un mensaje de bienvenida amigable para estudiantes sin experiencia
print("\n¡Hola! 👋 Bienvenido a tu Calculadora Financiera de Divisas")
print("Te ayudaré a convertir Pesos Colombianos a Dólares y Euros\n")

# Explicar brevemente qué hace el programa (educativo)
print("📚 Este programa te enseñará:")
print("   • Cómo capturar datos del usuario")
print("   • Convertir texto a números (conversión de tipos)")
print("   • Realizar cálculos financieros básicos")
print("   • Mostrar resultados de forma profesional\n")

# ===============================================================================
# 2. ENTRADAS DE DATOS (INPUTS & CASTING)
# ===============================================================================

# Envolvemos todo en un bloque try-except para capturar errores
try:
    # Solicitamos el monto en Pesos Colombianos
    # input() detiene el programa y espera que el usuario escriba algo
    monto_cop_texto = input("¿Cuántos Pesos Colombianos (COP) deseas convertir? ")
    
    # float() convierte el texto que escribió el usuario en un número decimal
    # Lo usamos en finanzas porque permite trabajar con centavos (0.50, 1.25, etc)
    monto_cop = float(monto_cop_texto)
    
    # Solicitamos la Tasa de Cambio del Dólar (TRM)
    # La TRM es el valor oficial del dólar estadounidense en Colombia
    trm_dolar_texto = input("¿Cuál es la TRM del Dólar hoy? (Ej: 4200.50) ")
    
    # Convertimos la TRM a número decimal (float) para hacer cálculos matemáticos
    trm_dolar = float(trm_dolar_texto)
    
    # Solicitamos la tasa del Euro
    # El Euro es la moneda oficial de la Unión Europea
    tasa_euro_texto = input("¿Cuál es la tasa del Euro? (Ej: 4600.75) ")
    
    # Convertimos la tasa del Euro a número decimal
    tasa_euro = float(tasa_euro_texto)
    
    print("\n✅ ¡Datos capturados correctamente!\n")

# Si el usuario ingresa letras en lugar de números, Python lanza un ValueError
except ValueError:
    # Mostramos un mensaje de error amigable con emojis
    print("\n" + "=" * 70)
    print("❌ ¡ERROR! Has ingresado letras en lugar de números")
    print("=" * 70)
    print("⚠️  Por favor, escribe solo números (sin letras ni caracteres especiales)")
    print("    Ejemplo correcto: 100000 o 4200.50\n")
    
    # Las siguientes líneas solo se ejecutan si hubo un error
    # Usamos 'raise' para que el programa muestre el error completo
    raise

# ===============================================================================
# 3. PROCESAMIENTO (CÁLCULOS)
# ===============================================================================

# Hacemos el cálculo para convertir COP a Dólares
# La fórmula es: Monto en dólares = Monto en COP / TRM del Dólar
monto_usd = monto_cop / trm_dolar

# Hacemos el cálculo para convertir COP a Euros
# La fórmula es: Monto en euros = Monto en COP / Tasa del Euro
monto_eur = monto_cop / tasa_euro

# ===============================================================================
# 4. SALIDA DE INFORMACIÓN (OUTPUTS)
# ===============================================================================

# Crear un encabezado visual atractivo con caracteres
print("=" * 70)
print("📊 RESULTADOS DE TU CONVERSIÓN 📊")
print("=" * 70)

# Mostrar el monto original en COP con formato profesional
# f-string: permite insertar variables dentro del texto
# :,.2f: significa "con separadores de miles y exactamente 2 decimales"
# Ejemplo: 1000000.00 se muestra como 1,000,000.00
print(f"\n💵 Monto Original: ${monto_cop:,.2f} COP")

# Mostrar una línea divisoria para mejorar la presentación visual
print("-" * 70)

# Mostrar resultado en Dólares
# 🇺🇸 es un emoji de bandera para que sea más visual
print(f"\n🇺🇸 Equivalente en Dólares: ${monto_usd:,.2f} USD")

# Mostrar un pequeño cálculo educativo (muestra cómo se calculó)
print(f"   (Cálculo: {monto_cop:,.2f} ÷ {trm_dolar:,.2f} = {monto_usd:,.2f})")

# Mostrar resultado en Euros
# 🇪🇺 es un emoji de bandera para que sea más visual
print(f"\n🇪🇺 Equivalente en Euros: €{monto_eur:,.2f} EUR")

# Mostrar un pequeño cálculo educativo (muestra cómo se calculó)
print(f"   (Cálculo: {monto_cop:,.2f} ÷ {tasa_euro:,.2f} = {monto_eur:,.2f})")

# Mostrar una línea final para cerrar elegantemente
print("\n" + "=" * 70)
print("✨ ¡Conversión completada exitosamente! ✨")
print("=" * 70)

# Mostrar información adicional educativa
print("\n📌 NOTAS IMPORTANTES:")
print("   • El símbolo $ representa dinero en general")
print("   • USD = Dólares Estadounidenses")
print("   • EUR = Euros")
print("   • Las tasas de cambio van a variar cada día")
print("   • Siempre verifica las tasas oficiales antes de hacer transacciones\n")
```
